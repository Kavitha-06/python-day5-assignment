```python
# sorted list arrangement
list1=[]
n=int(input("enter the value"))
for i in range(0,n):
    a=int(input())
    list1.append(a)
list2=[]
m=list1.count(0)
for i in range(len(list1)):
    if list1[i]!=0:
        list2.append(list1[i])
list2.sort()
for i in range(m):
    list2.append(0)
print("the sorted list:"+str(list2))
```

```python
#merge two sorted list
list1=[]
n=int(input("enter the value"))
for i in range(0,n):
    a=int(input())
    list1.append(a)
list2=[]
n=int(input("enter the value"))
for i in range(0,n):
    a=int(input())
    list2.append(a)
res=sorted(list1+list2)
print("the sorted list is"+ str(res))
```
